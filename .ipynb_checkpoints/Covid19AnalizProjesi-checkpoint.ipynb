```python
import pandas as pd
import numpy as np

# 97 gibi anlamsız değerleri NaN yapalım
df_cleaned = df.replace(97, np.nan)

# '9999-99-99' olan ölüm tarihlerini NaN yapıp, IS_DEAD kolonu oluşturalım
df_cleaned['IS_DEAD'] = np.where(df_cleaned['ÖLÜM_TARİHİ'] == '9999-99-99', 0, 1)

# Eğer tarih tipine çevirmek istersen (gerekirse)
df_cleaned['ÖLÜM_TARİHİ'] = pd.to_datetime(df_cleaned['ÖLÜM_TARİHİ'], errors='coerce')

# Verinin ilk haline bakalım
print(df_cleaned.head())
```

```python
import numpy as np

# 1. 97 değerlerini kalıcı olarak NaN yapalım
df.replace(97, np.nan, inplace=True)

# 2. 'ÖLÜM_TARİHİ' kolonunda 9999-99-99 olanlara göre yeni bir IS_DEAD kolonu oluşturalım
df['IS_DEAD'] = np.where(df['ÖLÜM_TARİHİ'].isna(), 0, 1)

# 3. İstersen 'ÖLÜM_TARİHİ' kolonunu datetime formatına çevirebiliriz (opsiyonel)
df['ÖLÜM_TARİHİ'] = pd.to_datetime(df['ÖLÜM_TARİHİ'], errors='coerce')

# 4. Temizlik sonrası ilk 20 satırı görelim
print(df.head(20))
```

```python
df.head(20)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Yaş gruplarını oluşturma (0-10, 10-20, 20-30, ...)
bins = [i for i in range(0, 110, 10)]  # 0'dan 100'e kadar 10'ar yıllık gruplar
labels = [f'{i}-{i+9}' for i in range(0, 100, 10)]  # 10 yaş aralıkları için etiketler
df['Yaş Grubu'] = pd.cut(df['YAŞ'], bins=bins, labels=labels, right=False)

# Ölüm oranını hesaplayalım (observed=True ekledik)
age_group_death = df.groupby('Yaş Grubu', observed=True)['IS_DEAD'].agg(lambda x: (x == 1).sum() / len(x))

# Grafik oluşturalım
plt.figure(figsize=(10,6))
age_group_death.plot(kind='bar', color='skyblue')
plt.title('Yaş Gruplarına Göre Ölüm Oranı')
plt.xlabel('Yaş Grubu')
plt.ylabel('Ölüm Oranı')
plt.xticks(rotation=45)
plt.show()
```

```python
# Entübe olan hastaları filtreleyelim
entube_hastalar = df[df['ENTÜBE'] == 1]

# Entübe olanların ölüm oranını hesaplayalım
entube_olum_orani = entube_hastalar['IS_DEAD'].mean()

# Sonuçları yazdıralım
print(f"Entübe olan hastaların ölüm oranı: {entube_olum_orani:.2f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Temel İstatistikler
# Cinsiyet ve Ölüm Oranı
cinsiyet_olum_orani = df.groupby('CİNSİYET', observed=True)['IS_DEAD'].mean()
print("Cinsiyetlere göre ölüm oranı:\n", cinsiyet_olum_orani)

# Yaş Gruplarına Göre Ölüm Oranı
age_group_death = df.groupby('Yaş Grubu', observed=True)['IS_DEAD'].mean()
print("\nYaş gruplarına göre ölüm oranı:\n", age_group_death)

# Diyabet, KOAH, Astım, Zatürre gibi hastalıklar ve Ölüm Oranı
diyabet_olum_orani = df.groupby('DİABETES', observed=True)['IS_DEAD'].mean()
koah_olum_orani = df.groupby('KOAH', observed=True)['IS_DEAD'].mean()
astim_olum_orani = df.groupby('ASTIM', observed=True)['IS_DEAD'].mean()
zatürre_olum_orani = df.groupby('ZATÜRRE', observed=True)['IS_DEAD'].mean()

# Sonuçları yazdıralım
print("\nDiyabetli hastaların ölüm oranı:\n", diyabet_olum_orani)
print("\nKOAH hastalarının ölüm oranı:\n", koah_olum_orani)
print("\nAstımlı hastaların ölüm oranı:\n", astim_olum_orani)
print("\nZatürre hastalığı olanların ölüm oranı:\n", zatürre_olum_orani)

# Entübe hastaların ölüm oranı
entube_hastalar_olum_orani = df[df['ENTÜBE'] == 1]['IS_DEAD'].mean()
print(f"\nEntübe hastaların ölüm oranı: {entube_hastalar_olum_orani:.2f}")

# Hastalıklar ve diğer faktörlerin birlikte analizi (Hastalıkların etkisi)
# Örneğin, diyabet ve entübe olma durumu ile ölüm oranı
diyabet_entube_olum_orani = df.groupby(['DİABETES', 'ENTÜBE'], observed=True)['IS_DEAD'].mean()
print("\nDiyabet ve entübe durumuna göre ölüm oranı:\n", diyabet_entube_olum_orani)

# Hastalıkların etkisini görselleştirelim
plt.figure(figsize=(10, 6))

# Diyabetli, KOAH, Astım, Zatürre hastalıklarının ölüm oranlarını karşılaştıralım
disease_death = pd.DataFrame({
    'Diyabet': diyabet_olum_orani,
    'KOAH': koah_olum_orani,
    'Astım': astim_olum_orani,
    'Zatürre': zatürre_olum_orani
})

disease_death.plot(kind='bar', color=['lightblue', 'salmon', 'lightgreen', 'orange'])
plt.title('Hastalık Durumlarına Göre Ölüm Oranı')
plt.xlabel('Hastalık Durumu')
plt.ylabel('Ölüm Oranı')
plt.xticks(rotation=0)
plt.legend(title='Hastalık Durumu')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Cinsiyet ve Ölüm Oranı
plt.figure(figsize=(10, 6))
cinsiyet_olum_orani.plot(kind='bar', color=['lightblue', 'salmon'])
plt.title('Cinsiyetlere Göre Ölüm Oranı')
plt.xlabel('Cinsiyet')
plt.ylabel('Ölüm Oranı')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 2. Yaş Gruplarına Göre Ölüm Oranı
plt.figure(figsize=(10, 6))
age_group_death.plot(kind='bar', color='lightgreen')
plt.title('Yaş Gruplarına Göre Ölüm Oranı')
plt.xlabel('Yaş Grubu')
plt.ylabel('Ölüm Oranı')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Hastalıklar ve Ölüm Oranı (Diyabet, KOAH, Astım, Zatürre)
disease_death.plot(kind='bar', figsize=(12, 8), color=['lightblue', 'salmon', 'lightgreen', 'orange'])
plt.title('Hastalık Durumlarına Göre Ölüm Oranı')
plt.xlabel('Hastalık Durumu')
plt.ylabel('Ölüm Oranı')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
